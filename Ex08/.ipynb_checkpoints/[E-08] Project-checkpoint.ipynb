### (3) Face Detection

- `dlib`Ïùò `face detector`Î•º ÏÇ¨Ïö©ÌïòÏó¨ ÏñºÍµ¥Ïùò ÏúÑÏπòÎ•º Ï∞æÏïÑ bounding boxÎ•º ÏÉùÏÑ±ÌïúÎã§.

```python
# detectorÎ•º ÏÑ†Ïñ∏
detector_hog = dlib.get_frontal_face_detector()
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1) # (image, num of image pyramid)
```

<br>

- `detector_hog`Ïùò Îëê Î≤àÏß∏ ÌååÎùºÎØ∏ÌÑ∞Îäî Ïù¥ÎØ∏ÏßÄ ÌîºÎùºÎØ∏ÎìúÏùò Ïàò
    - `Ïù¥ÎØ∏ÏßÄ ÌîºÎùºÎØ∏Îìú`ÎûÄ ÎèôÏùºÌïú Ïù¥ÎØ∏ÏßÄÏóê ÎåÄÌïú Îã§ÏñëÌïú ÏÇ¨Ïù¥Ï¶à set
    - Ïù¥ÎØ∏ÏßÄÎ•º upsampling Î∞©Î≤ïÏùÑ ÌÜµÌï¥ ÌÅ¨Í∏∞Î•º ÌÇ§Ïö∞Îäî Í≤ÉÏùÑ Ïù¥ÎØ∏ÏßÄ ÌîºÎùºÎØ∏ÎìúÎùºÍ≥† ÌïúÎã§.
    - Ïù¥ÎØ∏ÏßÄ ÌîºÎùºÎØ∏ÎìúÏóêÏÑú ÏñºÍµ¥ÏùÑ Îã§Ïãú Í≤ÄÏ∂úÌïòÎ©¥ ÏûëÍ≤å Ï¥¨ÏòÅÎêú ÏñºÍµ¥ÏùÑ ÌÅ¨Í≤å Î≥º Ïàò ÏûàÍ∏∞ ÎïåÎ¨∏Ïóê Ï†ïÌôïÌïú Í≤ÄÏ∂úÏù¥ Í∞ÄÎä•.<br><br>
    
> ‚òÜ UpsamplingÏù¥ÎûÄ?

```python
print(dlib_rects)   

# Ïó¨Í∏∞ÏÑú  345 = left / 98 = top / 531 = right / 284 = bottom
```

```python
color = [(255,0,0), (0,255,0), (0,0,255), (255,255,0), (255,0,255)] # Ìåå Ï¥à Îπ® Ï≤≠Î°ù Î≥¥

for i, dlib_rect in enumerate(dlib_rects):
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()
    
    # cv2.rectangle(img_show, (l, t), (r, b), (255, 0, 0), 2, lineType=cv2.LINE_AA)
    cv2.rectangle(img_show, (l, t), (r, b), color[i], 2, lineType=cv2.LINE_AA)
```

```python
img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

### (4) ÏñºÍµ¥ ÎûúÎìúÎßàÌÅ¨ face landmark

- ÏñºÍµ¥Ïùò Ïù¥Î™©Íµ¨ÎπÑ ÏúÑÏπòÎ•º Ï∂îÎ°†ÌïòÎäî Í≤ÉÏùÑ **`face landmark localization`** Í∏∞Ïà†Ïù¥ÎùºÍ≥† ÌïúÎã§.<br>
- face landmarkÎäî detectionÏùò Í≤∞Í≥ºÎ¨ºÏù∏ bounding boxÎ°ú ÏûòÎùºÎÇ∏(crop) ÏñºÍµ¥ Ïù¥ÎØ∏ÏßÄÎ•º Ïù¥Ïö©ÌïúÎã§.

face landmarkÏôÄ Í∞ôÏù¥ **Í∞ùÏ≤¥ ÎÇ¥Î∂ÄÏùò Ï†ê**ÏùÑ Ï∞æÎäî Í∏∞Ïà†ÏùÑ **object keypoint estimation**Ïù¥ÎùºÍ≥† ÌïúÎã§.
- 1) top down
- 2) bottom up

Dlib landmark localization
- ibug 300-W Îç∞Ïù¥ÌÑ∞ÏÖã >> pretrained model

```python
# Ïª§ÎÑêÎ°ú Ï†ÄÏû•Ìïú labdnmark Î™®Îç∏ÏùÑ Î∂àÎü¨Ïò®Îã§.

model_path = os.getenv('HOME')+'/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)
```

```python
# landmark_predictorÎäî RGBÏù¥ÎØ∏ÏßÄÏôÄ dlib.rectangleÏùÑ ÏûÖÎ†•Î∞õÏïÑ dlib.full_object_detectionÎ•º Î∞òÌôòÌïúÎã§.

list_landmarks =[]

# ÏñºÍµ¥ ÏòÅÏó≠ Î∞ïÏä§ ÎßàÎã§ face landmarkÎ•º Ï∞æÏïÑÎÇ∏Îã§.
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    # face landmark Ï¢åÌëúÎ•º Ï†ÄÏû•Ìï¥ÎëîÎã§.
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)
    
print(len(list_landmarks[0]))
```

```python
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 2, (0, 255, 255), -1)
         # -1 :: thickness - ÏÑ†Ïùò ÎëêÍªò, -1Ïù¥Î©¥ Ïõê ÏïàÏ™ΩÏùÑ Ï±ÑÏõÄ
        
img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

### (5) Ïä§Ìã∞Ïª§ Ï†ÅÏö©ÌïòÍ∏∞

<br>

**Ïù¥ÎØ∏ÏßÄ Í≤ΩÎ°ú Î∂àÎü¨Ïò§Í∏∞**

```python
sticker_path = os.getenv('HOME')+'/aiffel/camera_sticker/images/king.png'
img_sticker = cv2.imread(sticker_path) # Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄÎ•º Î∂àÎü¨ÏòµÎãàÎã§
```

**Ï¢åÌëú ÌôïÏù∏**

```python
def sticker_positioning(dlib_rects, list_landmarks, idx):
    
    
    # Ïä§Ìã∞Ïª§ Î∂àÎü¨Ïò§Í∏∞
    img_sticker = cv2.imread(sticker_path) # Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄÎ•º Î∂àÎü¨ÏòµÎãàÎã§
    
    # index Î≥Ñ dlib_rectsÏôÄ list_landmarks
    dlib_rect = dlib_rects[idx]
    landmark = list_landmarks[idx]

    x = landmark[30][0] # ÏΩîÏùò x Ï¢åÌëú
    y = landmark[30][1] - dlib_rect.height()//2 # ÏΩîÏùò yÏ¢åÌëúÏóêÏÑú bounding box ÎÜíÏù¥Ïùò Ï†àÎ∞òÎßåÌÅº ÎπºÏ£ºÍ∏∞
    w = h = dlib_rect.width() # Ïù¥ÎØ∏ÏßÄ ÏÇ¨Ïù¥Ï¶àÎ•º bounding boxÏùò Ìè≠Ïùò Í∏∏Ïù¥Î°ú ÏÑ§Ï†ï
    
    img_sticker = cv2.resize(img_sticker, (w,h))  # Ïù¥ÎØ∏ÏßÄ ÏÇ¨Ïù¥Ï¶à ÏÑ§Ï†ï
    
    # ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏóê Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄÎ•º Ï∂îÍ∞ÄÌïòÍ∏∞ ÏúÑÌï¥ÏÑú x, y Ï¢åÌëúÎ•º Ï°∞Ï†ï, Ïù¥ÎØ∏ÏßÄ ÏãúÏûëÏ†êÏùÄ top-left Ï¢åÌëú
    refined_x = x - w // 2  
    refined_y = y - h
    
    # 1) Ïä§Ìã∞Ïª§Ïùò ÏãúÏûëÏ†êÏù¥ ÏñºÍµ¥ ÏÇ¨ÏßÑÏùò ÏòÅÏó≠ÏùÑ Î≤óÏñ¥ÎÇòÎ©¥ ÏùåÏàòÎ°ú ÌëúÌòÑÎê®
    # 2) Ïä§Ìã∞Ïª§Ïùò ÏãúÏûëÏ†ê + Ïä§Ìã∞Ïª§ ÏÇ¨ÏßÑÏùò ÌÅ¨Í∏∞Í∞Ä ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏùò ÌÅ¨Í∏∞Î≥¥Îã§ ÌÅ∞ Í≤ΩÏö∞
    if refined_x < 0: 
        img_sticker = img_sticker[:, -refined_x:]
        refined_x = 0
    if refined_y < 0:
        img_sticker = img_sticker[-refined_y:, :]
        refined_y = 0
        
    sticker_area = img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
    img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
        np.where(img_sticker==0,sticker_area,img_sticker).astype(np.uint8)
    
    sticker_area = img_bgr[refined_y:refined_y +img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
    img_bgr[refined_y:refined_y +img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
        np.where(img_sticker==0,sticker_area,img_sticker).astype(np.uint8)
```

```python
for i in range(5):
    sticker_positioning(dlib_rects, list_landmarks, i)
```

```python
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```

**Ï¢åÌëú ÌôïÏù∏**

```python
def sticker_positioning(dlib_rects, list_landmarks, idx):
    
    
    # Ïä§Ìã∞Ïª§ Î∂àÎü¨Ïò§Í∏∞
    img_sticker = cv2.imread(sticker_path) # Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄÎ•º Î∂àÎü¨ÏòµÎãàÎã§
    
    # index Î≥Ñ dlib_rectsÏôÄ list_landmarks
    dlib_rect = dlib_rects[idx]
    landmark = list_landmarks[idx]
    
    print(dlib_rect, dlib_rect.width(), dlib_rect.height())

    x = landmark[30][0] # ÏΩîÏùò x Ï¢åÌëú
    y = landmark[30][1] # ÏΩîÏùò yÏ¢åÌëúÏóêÏÑú bounding box ÎÜíÏù¥Ïùò Ï†àÎ∞òÎßåÌÅº ÎπºÏ£ºÍ∏∞
    w = h = dlib_rect.width() # Ïù¥ÎØ∏ÏßÄ ÏÇ¨Ïù¥Ï¶àÎ•º bounding boxÏùò Ìè≠Ïùò Í∏∏Ïù¥Î°ú ÏÑ§Ï†ï
    
    img_sticker = cv2.resize(img_sticker, (w,h))  # Ïù¥ÎØ∏ÏßÄ ÏÇ¨Ïù¥Ï¶à ÏÑ§Ï†ï
    
    roi = img_bgr[dlib_rect.top():dlib_rect.top()+dlib_rect.width(),   dlib_rect.left():dlib_rect.left()+dlib_rect.width()]
    
    print('Ïù¥ÎØ∏ÏßÄ ÏÇ¨Ïù¥Ï¶à Ï∂úÎ†• :',img_sticker.shape)
    print('roi ÏÇ¨Ïù¥Ï¶à Ï∂úÎ†• :', roi.shape)
    print('\n=================================\n')
    
    # ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏóê Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄÎ•º Ï∂îÍ∞ÄÌïòÍ∏∞ ÏúÑÌï¥ÏÑú x, y Ï¢åÌëúÎ•º Ï°∞Ï†ï, Ïù¥ÎØ∏ÏßÄ ÏãúÏûëÏ†êÏùÄ top-left Ï¢åÌëú
    refined_x = x - w // 2  
    refined_y = y - h // 2
    
    # 1) Ïä§Ìã∞Ïª§Ïùò ÏãúÏûëÏ†êÏù¥ ÏñºÍµ¥ ÏÇ¨ÏßÑÏùò ÏòÅÏó≠ÏùÑ Î≤óÏñ¥ÎÇòÎ©¥ ÏùåÏàòÎ°ú ÌëúÌòÑÎê®
    # 2) Ïä§Ìã∞Ïª§Ïùò ÏãúÏûëÏ†ê + Ïä§Ìã∞Ïª§ ÏÇ¨ÏßÑÏùò ÌÅ¨Í∏∞Í∞Ä ÏõêÎ≥∏ Ïù¥ÎØ∏ÏßÄÏùò ÌÅ¨Í∏∞Î≥¥Îã§ ÌÅ∞ Í≤ΩÏö∞
    if refined_x < 0: 
        img_sticker = img_sticker[:, -refined_x:]
        refined_x = 0
    if refined_y < 0:
        img_sticker = img_sticker[-refined_y:, :]
        refined_y = 0
        
#     sticker_area = img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
#     img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
#         np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)
#     post_processing = cv2.addWeighted(roi, 0.5, img_sticker, 0.5, 0)
#     img_show[dlib_rect.top():dlib_rect.top()+dlib_rect.width(),   dlib_rect.left():dlib_rect.left()+dlib_rect.width()] = post_processing

    
    sticker_area = img_bgr[refined_y:refined_y +img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
    img_bgr[refined_y:refined_y +img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
        np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)
    post_processing = cv2.addWeighted(roi, 1, img_sticker, 0.1, 0)
    img_bgr[dlib_rect.top():dlib_rect.top()+dlib_rect.width(),   dlib_rect.left():dlib_rect.left()+dlib_rect.width()] = post_processing
```
